```python
import numpy as np

current = -1
def RoundRobin(queues):
    global current
    current = (current+1)%num_queues
    return current

def ShortestFirst(queues):
    return queues.index(min(queues))

def ChooseQueue(queues):
    return ShortestFirst()

def simulate(chooseQueue,
            num_queues,
            total_people_in_queue,
            average_arrival_rate_main,
            average_departure_rate_total,
            printFlag = False):

    # Initialize queues with empty queue sizes
    queues = [0] * num_queues

    total_people_departed = 0  # Initialize total people who have departed
    total_time_minutes = 0  # Initialize total time

    def GetArrivalsToMainQueueThisMinute():
        arrivals_main_this_minute = 0
        remain = total_people_in_queue - (sum(queues) + total_people_departed)

        if sum(queues) + total_people_departed < total_people_in_queue:
            arrivals_main_this_minute = min([np.random.poisson(average_arrival_rate_main),remain])

        return arrivals_main_this_minute

    def DivideArrivals(total):
        arrivals_this_minute = [0] * num_queues
        for i in range(total):
            queue_index = chooseQueue(queues)
            arrivals_this_minute[queue_index] += 1
            queues[queue_index] += 1
        return arrivals_this_minute

    def DepartFromQueues():
        departures_this_minute = [0] * num_queues
        # Check if each queue is not empty before generating departures
        for i in range(num_queues):
            if queues[i] > 0:
                departures_this_minute[i] = min(np.random.poisson(average_departure_rate_total / num_queues), queues[i])
                queues[i] -= departures_this_minute[i]
        return departures_this_minute

    while total_people_departed < total_people_in_queue:

        arrivals_main_this_minute = GetArrivalsToMainQueueThisMinute()
        arrivals_this_minute = DivideArrivals(arrivals_main_this_minute)
        departures_this_minute = DepartFromQueues()

        total_people_departed += sum(departures_this_minute)
        total_time_minutes += 1

#         print(f"Minute {total_time_minutes}:")
#         print (f"Arrived {sum(arrivals_this_minute)} {arrivals_main_this_minute}")
#         print(f"Depart {sum(departures_this_minute)}")
#         print(f"total_people_departed {total_people_departed}:")
#         for j in range(num_queues):
#             print(f"Queue {j + 1} size: {queues[j]}")  # Ensure non-negative values
#             print(f"Arrival to Queue {j + 1}: {arrivals_this_minute[j]}")
#             print(f"Departures from Queue {j + 1}: {departures_this_minute[j]}")
#         print("\n")
    if printFlag:
        print(f"Total time to process {total_people_in_queue} people: {total_time_minutes} minutes")
    return total_time_minutes;

num_queues = 10  # Number of queues
total_people_in_queue = 500  # Total number of people in the queue
average_arrival_rate_main = 100 / 60.0  # Average arrivals per minute at the main queue
average_departure_rate_total = 80 / 60.0  # Average total departures per minute
RR = []
SF = []
for i in range (1):
    RR.append(simulate(RoundRobin,
                      num_queues,
                      total_people_in_queue,
                      average_arrival_rate_main,
                      average_departure_rate_total))
    SF.append(simulate(ShortestFirst,
                      num_queues,
                      total_people_in_queue,
                      average_arrival_rate_main,
                      average_departure_rate_total))
RRavg = sum(RR)/len(RR)
SFavg = sum(SF)/len(SF)

print(f"RR: {sum(RR)/len(RR)}")
print(f"SF: {sum(SF)/len(SF)}")
print(f"diff {RRavg-SFavg} prec {RRavg/SFavg * 100-100}% diff")
```

```python
import tkinter as tk
from tkinter import messagebox
import firebase_admin
from firebase_admin import credentials, db

def InitFirebase():
    # Initialize Firebase Admin SDK with your JSON key file
    cred = credentials.Certificate('sq.json')
    if firebase_admin._DEFAULT_APP_NAME not in firebase_admin._apps:
        firebase_admin.initialize_app(cred, {
            'databaseURL': 'https://smartqueue-60f8d-default-rtdb.europe-west1.firebasedatabase.app'
        })
        
def SetLines(data, size):
    res = []
    refStr = "UsersData/ONdzE675Ynf9M0yZ4IsIxMjkEVv1/readings/"
    for i in range(size):
        lineRef = refStr + str(i)
        ref = db.reference(lineRef)
        updatedData = ref.get()
        updatedData['num_of_people'] = data[i]
        ref.set(updatedData)
    return res

class QueueApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Queue App")
        self.root.geometry("800x600")  # Set initial window size

        self.queue_count = 0
        self.queue_entries = []
        self.queue_labels = []
        self.queue_frame = None
        self.send_button = None
        self.message_label = None

        self.create_widgets()

    def create_widgets(self):
        self.label = tk.Label(self.root, text="Enter the number of queues:")
        self.label.pack()

        self.queue_count_entry = tk.Entry(self.root)
        self.queue_count_entry.pack()

        self.submit_button = tk.Button(self.root, text="Submit", command=self.create_queue_entries)
        self.submit_button.pack()

    def create_queue_entries(self):
        try:
            new_queue_count = int(self.queue_count_entry.get())
            if new_queue_count < 1:
                messagebox.showerror("Error", "Number of queues should be greater than 0")
                return

            # Clear previous labels, entries, and message
            if self.queue_frame:
                self.queue_frame.destroy()
            if self.send_button:
                self.send_button.destroy()
            if self.message_label:
                self.message_label.destroy()

            self.queue_labels.clear()
            self.queue_entries.clear()

            # Create new frame for queue entry boxes
            self.queue_frame = tk.Frame(self.root)
            self.queue_frame.pack()

            # Create new queue entry boxes
            for i in range(new_queue_count):
                label = tk.Label(self.queue_frame, text=f"Queue {i+1}:")
                label.pack()
                self.queue_labels.append(label)

                entry = tk.Entry(self.queue_frame)
                entry.pack()
                self.queue_entries.append(entry)

            self.queue_count = new_queue_count

            # Create "Send Queues" button
            self.send_button = tk.Button(self.root, text="Send Queues", command=self.send_queues)
            self.send_button.pack()
        except ValueError:
            messagebox.showerror("Error", "Please enter a valid integer.")

    def send_queues(self):
        queues = []
        for entry in self.queue_entries:
            queue_value = int(entry.get())
            if queue_value < 0:
                messagebox.showerror("Error", "Number of queues should be greater than 0")
            queues.append(queue_value)

        # Clear the queue frame and send button
        self.queue_frame.destroy()
        self.send_button.destroy()
        
        SetLines(queues, len(queues))

        # Display "Sent successfully" message
        self.message_label = tk.Label(self.root, text="Sent successfully!")
        self.message_label.pack()
        
        # Print the queues to the terminal
        print("Queues:", queues)

if __name__ == "__main__":
    InitFirebase()
    root = tk.Tk()
    app = QueueApp(root)
    root.mainloop()
```
